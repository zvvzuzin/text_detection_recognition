```python
import cv2
import matplotlib.pyplot as plt
from text_recognition import Container_Number_Recognition
```

```python
xml_file = 'text-detection.xml'
bin_file = 'text-detection.bin'
cont_num_rec = Container_Number_Recognition(xml_file, bin_file)
```

```python
frame = cv2.imread('image.jpg')
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
cont_rects = [[400, 470, 870, 920],
              [0, 0, 380, 400],
              [0, 470, 400, 920],
              [380, 0, 850, 420],
              [400, 920, 850, 1280]
             ]
plt.figure(figsize=(20,10))
for i, rect in enumerate(cont_rects):
    plt.subplot(1, len(cont_rects), i+1)
    plt.imshow(frame[rect[0]:rect[2], rect[1]:rect[3]])
```

```python
num_rects = cont_num_rec.detect_numbers(frame, cont_rects)
nums = cont_num_rec.recognize_numbers(frame, num_rects)
plt.figure(figsize=(20,10))
for i, (rect, num) in enumerate(zip(num_rects, nums)):
    plt.subplot(1, len(num_rects), i+1)
    plt.imshow(frame[rect[0]:rect[2], rect[1]:rect[3]])
    plt.title(num)
```

```python
nums = cont_num_rec.predict(frame, cont_rects)
print(nums)
```
